```python

import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from pathlib import Path
from astropy.io import fits
import numpy as np
```

```python
datPath = Path("C:/Users/ajo5182/Documents/astro/psf/psf_source_localization/data/test.npz")
data = np.load(datPath, allow_pickle=True)
```

```python
xSigma = np.array(data['xSigma'])
ySigma = np.array(data['ySigma'])
peakSig = np.array(data['peakSig'])
```

```python
xSigma = xSigma[peakSig != np.array(None)]
ySigma = ySigma[peakSig != np.array(None)]
peakSig = peakSig[peakSig != np.array(None)]
```

```python
xfwhm = 2.355*xSigma
yfwhm = 2.355*ySigma

xlocalization = 60*xfwhm/peakSig # localization in arcmin
ylocalization = 60*yfwhm/peakSig # localization in arcmin
```

```python
plt.hist(xlocalization)
plt.hist(ylocalization)
```
